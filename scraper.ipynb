```python
import requests
import pandas as pd
import json
from bs4 import BeautifulSoup

```

```python
url = 'https://octane.gg/stats/players?mode=3&minGames=50&group=rlcs2122fall'
html = requests.get(url)
soup = BeautifulSoup(html.content, "html.parser")
```

```python
# testing scrapage
# op = open('output.html', 'w')
# op.write(soup.prettify())
# op.close()
```

```python
stat_json = soup.find_all('script')[1].text.strip()

df = json.loads(stat_json)

nested = pd.json_normalize(df['props']['pageProps'], record_path=['stats'])
```

```python
clean_stats = nested.drop(['events', 'teams', 'opponents', 'startDate', 'endDate',
             'player._id','player.slug','games.replays','games.replaySeconds',
             'matches.replays','player.team._id','player.team.slug','player.team.name',
             'player.team.region','player.team.image','player.accounts',
             'player.relevant','player.name','player.team.relevant'], axis=1)

clean_stats
```
